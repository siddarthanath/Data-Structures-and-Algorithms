## Intuition (Theoretical)

1. Start off with two pointers (one at the start and one at the end).
2. Find the half-way point of these pointers.
3. If the value you seek is less than the value at this half-way point, we move the pointer at the end to become the pointer that sits directly on the left of the half-way value. Else, the value we seek is greater than the value at this half-way point so we move the pointer at the start to become the pointer that sits directly on the right of the half-way value.
4. Repeat Steps 2-3 till you find the value you want.

## Code

```python
# Library imports
from typing import List, Union
```

```python
# Binary Search Algorithm


class BinarySearch:
    def __init__(
        self, input_list: List[Union[float, int, str]], value: Union[float, int, str]
    ):
        """This initialises general values which will be used repeatedly in the class."""
        self.original_list = input_list
        self.value = value

    def check_list(self):
        """Check for edge cases for lists."""
        # Check if list is empty
        if len(self.original_list) == 0:
            raise ValueError("List is empty!")

    def execute(self):
        """This performs linear search."""
        # Check list compatibility
        self.check_list()
        # Set the two pointers (as index)
        start_pointer = 0
        end_pointer = len(self.original_list)-1
        # Store counter for complexity
        count = 1
        # Condition to meet is that the value is found
        while start_pointer <= end_pointer:
            # Set the mid pointer
            mid_pointer = (start_pointer + end_pointer) // 2
            print(f"- Iteration {count}...")
            # Check if the value of the mid pointer is greater than the value we seek
            if self.original_list[mid_pointer] == self.value:
                print(f"The value is in position: {mid_pointer}")
                return 
            elif self.original_list[mid_pointer] > self.value:
                 # Increment
                count += 1
                # Move start pointer
                end_pointer = mid_pointer - 1
            else:
                 # Increment
                count += 1
                # Move end pointer
                start_pointer = mid_pointer + 1
        print("Value does not exist in list!")
        return
```

## Test

```python
# Instatiate class object
kwargs_1 = {"input_list": [3, 6, 7, 9, 11, 16], "value": 9}
bs_alg_1 = BinarySearch(**kwargs_1)
# Perform binary search
bs_alg_1.execute()
```
