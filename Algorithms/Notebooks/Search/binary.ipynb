## Intuition (Theoretical)

1. Start off with two pointers (one at the start and one at the end).
2. Find the half-way point of these pointers.
3. If the value you seek is less than the value at this half-way point, we move the pointer at the end to become the pointer that sits directly on the left of the half-way value. Else, the value we seek is greater than the value at this half-way point so we move the pointer at the start to become the pointer that sits directly on the right of the half-way value.
4. Repeat Steps 2-3 till you find the value you want.

## Code

```python
# Library imports
from typing import List, Union
```

```python
"""This function performs binary search."""

def binary_search(input_list: List[Union[float, int]], value: Union[float, int]) -> List[Union[float, int]]:
    # Set the two pointers (as index)
    start_pointer = 0
    end_pointer = len(input_list)-1
    # Condition to meet is that the value is found
    print("Initiate Binary Search... \n")
    while start_pointer <= end_pointer:
        # Set the mid pointer
        mid_pointer = (start_pointer + end_pointer) // 2
        # Check if the value of the mid pointer is greater than the value we seek
        if input_list[mid_pointer] == value:
            print(f"The value is in position: {mid_pointer} \n")
            print("Binary Search complete!")
            return 
        elif input_list[mid_pointer] > value:
                # Increment
            # Move start pointer
            end_pointer = mid_pointer - 1
        else:
            # Move end pointer
            start_pointer = mid_pointer + 1
    print("Value does not exist in list... \n")
    print("Binary Search complete!")
```

## Test

```python
# Value exists
input_list = [23, 79, 92, 13, 7]
value = 79
binary_search(input_list=input_list, value=value)
```

```python
# Value does not exists
input_list = [23, 79, 92, 13, 7]
value = 80
binary_search(input_list=input_list, value=value)
```
