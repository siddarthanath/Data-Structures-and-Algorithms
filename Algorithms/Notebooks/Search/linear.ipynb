## Intuition (Theoretical)

1. We start from the beginning of the list and we check sequentially (through each index), whether the value we want is in that position.
2. If it is, we either return True or return the index of that value.
3. Else, we return False (or -1).

## Code

```python
# Library imports
from typing import List, Union
```

```python
"""This function performs linear search."""


def linear_search(
    input_list: List[Union[float, int]], value: Union[float, int]
) -> List[Union[float, int]]:
    # Loop through entire list to check if value exists
    print("Initiate Linear Search... \n")
    for i in range(len(input_list)):
        # Check existence
        if input_list[i] == value:
            # Return the index of the value (NOTE: You can also return True, if you do not care about the position.)
            print(f"The value is in position: {i} \n")
            print("Linear Search complete!")
            # NOTE: The non-optimal search would go to the end of the list and not break early if the value is found!
            return
    # If value does not exist (NOTE: You can also return False.)
    print("Value does not exist in list... \n")
    print("Linear Search complete!")
```

## Test

```python
# Value exists
input_list = [23, 79, 92, 13, 7]
value = 79
linear_search(input_list=input_list, value=value)
```

```python
# Value does not exists
input_list = [23, 79, 92, 13, 7]
value = 80
linear_search(input_list=input_list, value=value)
```
