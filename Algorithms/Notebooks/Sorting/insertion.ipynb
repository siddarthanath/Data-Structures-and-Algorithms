## Intuition (Theoretical)

1. Take a static pointer at the start of the list. Take a variable pointer, which starts one index after the static pointer. Insert the value of variable pointer before or after the static pointer (depending on the sorting criteria).
2. This means the first two elements are sorted. Now increment the variable pointer and repeat Step 1 (but now including the check against the first two elements and so on).
3. This means that we are sorting the left-most array on every iteration (unlike selection sort, which simply finds the minimum value on the right array and then applies a swap).

## Code

````python
# Library imports
from typing import List, Union
````

````python
"""This function performs insertion sort."""


def insertion_sort(input_list: List[Union[float, int]]) -> List[Union[float, int]]:
    # Set iterable pointer
    print("Initiate Insertion Sort... \n")
    for i in range(1, len(input_list)):
        print(f"- Iteration {i}: {input_list} \n")
        # Apply insert
        insert(input_list[i], input_list, i)
    print(f"- Iteration {i+1}: {input_list} \n")
    print("Selection Sort complete!")


def insert(
    current_val: Union[float, int], input_list: List[Union[float, int]], index: int
):
    # Set iterable pointer
    for j in range(index - 1, -1, -1):
        # Check if current value is larger than indexed value in list
        if current_val >= input_list[j]:
            # Apply insertion
            input_list[j + 1] = current_val
            return
        # Replace value (NOTE: This is similar to selection sort as a "swap".)
        input_list[j + 1] = input_list[j]
    # If all values are being replaced, then the current value belongs at the start
    input_list[0] = current_val
````

## Test

````python
# Ascending Ordered List
input_list = [23, 79, 92, 13, 7]
insertion_sort(input_list=input_list)
````

- The code returns an ascending ordered list - for a descending ordered list, simply change ```if current_val >= input_list[j]``` to ```if current_val <= input_list[j]```. 
- Additionally, to return the modified lists, add returns to the end of the functions rather than print.
- In general, to familiarise yourself fully with the code, it is helpful to debug through each iteration to understand what is going on within the list!
